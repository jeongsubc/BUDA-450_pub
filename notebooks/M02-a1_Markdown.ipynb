# HTML 

Many similar and advanced syntaxs can be made through HTML.

## Heading
<h1>H1</h1>
<h2>H2</h2>
<h3>H3</h3>

## Emphasis
<em>Italic</em>
<strong>Bold</strong>
<strike>Strikethrough</strike>
<strong><em>Bold italic</em><strong>
<mark>Mark</mark>

## Lists

Ordered list
<ol>
    <li>Item 1</li>
    <li>Item 2</li>
    <li>Item 3</li>
</ol>

Unordered list
<ul>
    <li>Item 1</li>
    <li>Item 2</li>
    <li>Item 3</li>
</ul>

## Code

<code> <a>Example of code</a> </code>
